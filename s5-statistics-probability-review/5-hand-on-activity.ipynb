# Hypothesis Testing: Comparing Average Session Time Between Two Groups

This notebook demonstrates how to perform hypothesis testing to compare the average session time between two groups using Python.

## Scenario and Hypotheses

Suppose we have two groups of users, Group A and Group B, and we want to determine if there is a significant difference in their average session times.

- **Null Hypothesis (H₀):** The average session time for Group A is equal to Group B.
- **Alternative Hypothesis (H₁):** The average session time for Group A is different from Group B.

````python
# Import required libraries
import numpy as np
import scipy.stats as stats
````

````python
# Generate sample session time data for two groups
np.random.seed(42)  # For reproducibility
group_a = np.random.normal(loc=30, scale=5, size=100)  # Group A: mean=30, std=5, n=100
group_b = np.random.normal(loc=32, scale=5, size=100)  # Group B: mean=32, std=5, n=100

print(f'Group A mean: {group_a.mean():.2f}, std: {group_a.std():.2f}')
print(f'Group B mean: {group_b.mean():.2f}, std: {group_b.std():.2f}')
````

````python
# Perform independent t-test
t_stat, p_value = stats.ttest_ind(group_a, group_b)

print(f'T-statistic: {t_stat:.4f}')
print(f'P-value: {p_value:.4f}')
````

## Interpretation of Results

- If the p-value is less than 0.05, we reject the null hypothesis and conclude that there is a significant difference in average session time between the two groups.
- If the p-value is greater than 0.05, we fail to reject the null hypothesis and conclude that there is no significant difference.

## How to Perform the Same Test in Excel

You can use the following formula in Excel:

```
=T.TEST(array1, array2, tails, type)
```

- `array1`: Data for Group A
- `array2`: Data for Group B
- `tails`: 2 (for two-tailed test)
- `type`: 2 (for two-sample equal variance)

````python
# Visualization: Distribution of Session Times for Both Groups
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,6))
sns.histplot(group_a, color='blue', label='Group A', kde=True, stat='density', bins=20, alpha=0.6)
sns.histplot(group_b, color='orange', label='Group B', kde=True, stat='density', bins=20, alpha=0.6)

plt.axvline(group_a.mean(), color='blue', linestyle='dashed', linewidth=2, label=f'Group A Mean: {group_a.mean():.2f}')
plt.axvline(group_b.mean(), color='orange', linestyle='dashed', linewidth=2, label=f'Group B Mean: {group_b.mean():.2f}')

plt.title('Session Time Distribution by Group')
plt.xlabel('Session Time')
plt.ylabel('Density')
plt.legend()
plt.show()
````

### Visualization Explanation

The plot above shows the distribution of session times for Group A and Group B, with dashed lines indicating the mean session time for each group. This helps visually compare the two groups and supports the results of the hypothesis test.
