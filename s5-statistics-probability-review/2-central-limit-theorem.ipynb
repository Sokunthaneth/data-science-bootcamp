# Central Limit Theorem Demonstration

This notebook demonstrates the Central Limit Theorem (CLT) using Python. The CLT states that the sampling distribution of the sample mean approaches a normal distribution as the sample size increases, regardless of the population's distribution (provided the population has a finite variance).

```python
# Import required libraries
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## 1. Population Distribution

Let's create a non-normal population distribution (exponential) to demonstrate the CLT.

```python
# Generate a non-normal (exponential) population
population = np.random.exponential(scale=2, size=100000)

# Plot the population distribution
plt.figure(figsize=(8, 4))
sns.histplot(population, bins=50, kde=True, color='skyblue')
plt.title('Population Distribution (Exponential)')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

## 2. Sampling and the Central Limit Theorem

We'll draw many samples of different sizes from the population and plot the distribution of their means.

```python
# Function to draw sample means
def sample_means(population, sample_size, n_samples=1000):
    means = [np.mean(np.random.choice(population, size=sample_size, replace=False)) for _ in range(n_samples)]
    return means

sample_sizes = [2, 5, 30, 100]
plt.figure(figsize=(16, 8))
for i, size in enumerate(sample_sizes, 1):
    means = sample_means(population, size)
    plt.subplot(2, 2, i)
    sns.histplot(means, bins=30, kde=True, color='orange')
    plt.title(f'Sample Size = {size}')
    plt.xlabel('Sample Mean')
    plt.ylabel('Frequency')
plt.tight_layout()
plt.suptitle('Sampling Distribution of the Mean for Different Sample Sizes', y=1.02)
plt.show()
```

## 3. Conclusion

As the sample size increases, the distribution of the sample means becomes more normal (bell-shaped), even though the original population is not normal. This is the essence of the Central Limit Theorem.
