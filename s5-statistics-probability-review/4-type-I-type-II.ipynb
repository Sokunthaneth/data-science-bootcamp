# Type I and Type II Errors in Hypothesis Testing

This notebook demonstrates the concepts of Type I and Type II errors in hypothesis testing.

## 1. Definitions

- **Type I Error (False Positive):** Rejecting the null hypothesis ($H_0$) when it is actually true. Probability is denoted by $\alpha$ (significance level).
- **Type II Error (False Negative):** Failing to reject the null hypothesis when the alternative hypothesis ($H_1$) is true. Probability is denoted by $\beta$.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

## 2. Visualizing Type I and Type II Errors

Let's visualize the errors using two normal distributions: one under $H_0$ and one under $H_1$.

```python
# Parameters
mu0 = 0   # Mean under H0
mu1 = 2   # Mean under H1
sigma = 1 # Standard deviation
alpha = 0.05 # Significance level

# Critical value for right-tailed test
critical_value = norm.ppf(1 - alpha, loc=mu0, scale=sigma)

x = np.linspace(-3, 5, 500)
y0 = norm.pdf(x, mu0, sigma)
y1 = norm.pdf(x, mu1, sigma)

plt.figure(figsize=(10,6))
plt.plot(x, y0, label='H0: mu=0', color='blue')
plt.plot(x, y1, label='H1: mu=2', color='red')

# Shade Type I error region (right tail of H0)
plt.fill_between(x, 0, y0, where=(x > critical_value), color='blue', alpha=0.3, label='Type I Error (alpha)')

# Shade Type II error region (left of critical under H1)
plt.fill_between(x, 0, y1, where=(x <= critical_value), color='red', alpha=0.3, label='Type II Error (beta)')

plt.axvline(critical_value, color='black', linestyle='--', label='Critical Value')
plt.title('Type I and Type II Errors')
plt.xlabel('Test Statistic')
plt.ylabel('Probability Density')
plt.legend()
plt.show()
```

## 3. Summary

- **Type I Error ($\alpha$):** Blue shaded area — rejecting $H_0$ when it is true.
- **Type II Error ($\beta$):** Red shaded area — failing to reject $H_0$ when $H_1$ is true.

Adjusting the significance level ($\alpha$) changes the balance between Type I and Type II errors.
