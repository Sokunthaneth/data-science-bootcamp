# Demonstrating Probability Distributions

This notebook demonstrates four common probability distributions: Normal, Binomial, Poisson, and Uniform. For each distribution, we will generate random samples, visualize the distribution, and briefly describe its properties.

```python
# Import required libraries
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## 1. Normal Distribution

The normal distribution is a continuous, bell-shaped distribution that is symmetric about the mean.

```python
# Generate normal distribution data
normal_data = np.random.normal(loc=0, scale=1, size=1000)

# Plot normal distribution
plt.figure(figsize=(7,4))
sns.histplot(normal_data, bins=30, kde=True, color='skyblue')
plt.title('Normal Distribution (mean=0, std=1)')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

## 2. Binomial Distribution

The binomial distribution is a discrete distribution describing the number of successes in a fixed number of independent Bernoulli trials.

```python
# Generate binomial distribution data
binomial_data = np.random.binomial(n=10, p=0.5, size=1000)

# Plot binomial distribution
plt.figure(figsize=(7,4))
sns.histplot(binomial_data, bins=range(0,12), kde=False, color='orange')
plt.title('Binomial Distribution (n=10, p=0.5)')
plt.xlabel('Number of Successes')
plt.ylabel('Frequency')
plt.show()
```

## 3. Poisson Distribution

The Poisson distribution is a discrete distribution describing the probability of a given number of events occurring in a fixed interval of time or space.

```python
# Generate poisson distribution data
poisson_data = np.random.poisson(lam=3, size=1000)

# Plot poisson distribution
plt.figure(figsize=(7,4))
sns.histplot(poisson_data, bins=range(0,12), kde=False, color='green')
plt.title('Poisson Distribution (lambda=3)')
plt.xlabel('Number of Events')
plt.ylabel('Frequency')
plt.show()
```

## 4. Uniform Distribution

The uniform distribution is a continuous distribution where all intervals of the same length are equally probable.

```python
# Generate uniform distribution data
uniform_data = np.random.uniform(low=0, high=1, size=1000)

# Plot uniform distribution
plt.figure(figsize=(7,4))
sns.histplot(uniform_data, bins=30, kde=True, color='purple')
plt.title('Uniform Distribution (0, 1)')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```
