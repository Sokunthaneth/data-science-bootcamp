# Linear Regression with scikit-learn
In this notebook, we will demonstrate how to perform linear regression using scikit-learn on a housing price dataset. We will output the regression coefficient, RMSE, R² score, and visualize the regression line.

```python
# Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

## Load the Dataset
Read the housing price data from the CSV file.

```python
# Load the dataset
file_path = '../s10-linear-logistic-regression/housing_price_dataset.csv'
data = pd.read_csv(file_path)
data.head()
```

## Prepare the Data
Separate the features and target variable for regression.

```python
# Prepare features and target
X = data[['house_sqm']]
y = data['price']
```

## Fit Linear Regression Model
Train a linear regression model using scikit-learn.

```python
# Fit the linear regression model
model = LinearRegression()
model.fit(X, y)

# Make predictions
y_pred = model.predict(X)
```

## Output Coefficient, RMSE, and R² Score
Display the regression coefficient, root mean squared error, and R² score.

```python
# Output results
coefficient = model.coef_[0]
intercept = model.intercept_
rmse = np.sqrt(mean_squared_error(y, y_pred))
r2 = r2_score(y, y_pred)

print(f"Coefficient: {coefficient:.4f}")
print(f"Intercept: {intercept:.4f}")
print(f"RMSE: {rmse:.4f}")
print(f"R² Score: {r2:.4f}")
```

## Visualize the Regression Line
Plot the data points and the fitted regression line.

```python
# Plot the regression line
plt.figure(figsize=(8, 5))
plt.scatter(X, y, color='blue', label='Actual Data')
plt.plot(X, y_pred, color='red', linewidth=2, label='Regression Line')
plt.xlabel('House Size (sqm)')
plt.ylabel('Price (thousands)')
plt.title('Linear Regression: House Size vs Price')
plt.legend()
plt.show()
```
