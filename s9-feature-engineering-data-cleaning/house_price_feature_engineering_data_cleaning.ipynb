This notebook demonstrates feature engineering and data cleaning techniques on the Ames Housing dataset. Tasks include:
- Detecting & handling missing values
- Encoding categorical variables
- Outlier detection
- Preparing a clean dataset for modeling

```python
# üì• Load the Ames Housing dataset
import pandas as pd

url = "https://raw.githubusercontent.com/ageron/handson-ml/master/datasets/housing/housing.csv"
df = pd.read_csv(url)
df.head()
```

## üîç Detecting Missing Values

```python
df.isnull().sum()
```

## üõ†Ô∏è Handling Missing Values

```python
# Fill total_bedrooms with median
df['total_bedrooms'].fillna(df['total_bedrooms'].median(), inplace=True)

df.isnull().sum()
```

## üî§ Encoding Categorical Variables

```python
# One-hot encode 'ocean_proximity'
df = pd.get_dummies(df, columns=['ocean_proximity'], drop_first=True)
df.head()
```

## üö® Outlier Detection: IQR Method

```python
# Use IQR to detect outliers in 'median_house_value'
Q1 = df['median_house_value'].quantile(0.25)
Q3 = df['median_house_value'].quantile(0.75)
IQR = Q3 - Q1
outliers = df[(df['median_house_value'] < Q1 - 1.5 * IQR) | (df['median_house_value'] > Q3 + 1.5 * IQR)]

print(f'Total outliers in median_house_value: {len(outliers)}')
```

```python
# Cap upper outliers (optional)
upper = Q3 + 1.5 * IQR
df['median_house_value'] = df['median_house_value'].apply(lambda x: min(x, upper))
```

## üíæ Save Cleaned Dataset

```python
df.to_csv('housing_cleaned.csv', index=False)
df.head()
```
