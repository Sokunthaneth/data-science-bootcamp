This notebook demonstrates essential plotting techniques using Matplotlib and Seaborn, including line plots, bar charts, customizations, distributions, and advanced visualizations.

## 1. Import Required Libraries
Import Matplotlib, NumPy, and Seaborn for data visualization.

```python
# 1: Import libraries
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

## 2. Line Plot
Create a line plot of the sine function using Matplotlib.

```python
# 2: Line plot
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.figure(figsize=(8, 4))
plt.plot(x, y, label='sin(x)', color='blue', linestyle='--', marker='o')
plt.title('Line Plot Example')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.legend()
plt.grid(True)
plt.show()
```

## 3. Bar Chart
Create a bar chart to compare values across categories.

```python
# 3: Bar chart
categories = ['A', 'B', 'C', 'D']
values = [10, 24, 36, 18]
plt.figure(figsize=(6, 4))
plt.bar(categories, values, color='orange', edgecolor='black')
plt.title('Bar Chart Example')
plt.xlabel('Category')
plt.ylabel('Value')
plt.show()
```

## 4. Customizing Plots
Customize plot appearance with titles, labels, legends, and grid styles.

```python
# 4: Customizing plots
plt.figure(figsize=(8, 4))
plt.plot(x, np.cos(x), label='cos(x)', color='green', linewidth=2)
plt.title('Customized Plot', fontsize=16, fontweight='bold')
plt.xlabel('x', fontsize=12)
plt.ylabel('cos(x)', fontsize=12)
plt.legend(loc='upper right')
plt.grid(True, linestyle=':', color='gray')
plt.show()
```

## 5. Seaborn Distributions
Visualize data distributions using Seaborn's histogram and KDE plot.

```python
# 5: Seaborn distributions
data = np.random.normal(loc=0, scale=1, size=1000)
plt.figure(figsize=(8, 4))
sns.histplot(data, kde=True, color='purple')
plt.title('Seaborn Distribution Plot')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

## 6. Styling and Aesthetics
Apply different Matplotlib styles for improved plot appearance.

```python
# 6: Styling and aesthetics
plt.style.use('ggplot')
plt.figure(figsize=(8, 4))
plt.plot(x, np.sin(x), label='sin(x)')
plt.plot(x, np.cos(x), label='cos(x)')
plt.title('Styled Plot with ggplot')
plt.xlabel('x')
plt.ylabel('Value')
plt.legend()
plt.show()
```

## Reset Style to Default
Reset Matplotlib style to the default settings.

```python
# Reset style to default
plt.style.use('default')
```

# Additional Matplotlib and Seaborn Functionalities

This section demonstrates more important plotting features, including scatter plots, subplots, annotations, saving figures, pie charts, histograms, error bars, log scales, 3D plotting, and advanced seaborn visualizations like pair plots, heatmaps, box plots, violin plots, swarm plots, FacetGrid, and regression plots.

## 7. Scatter Plot
Create a scatter plot to visualize the relationship between two variables.

```python
# 7: Scatter plot
np.random.seed(0)
x_scatter = np.random.rand(50)
y_scatter = np.random.rand(50)
plt.figure(figsize=(6, 4))
plt.scatter(x_scatter, y_scatter, color='teal', edgecolor='black')
plt.title('Scatter Plot Example')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

## 8. Subplots
Display multiple plots side by side using subplots.

```python
# 8: Subplots
fig, axs = plt.subplots(1, 2, figsize=(10, 4))
axs[0].plot(x, np.sin(x), color='blue')
axs[0].set_title('Sin(x)')
axs[1].plot(x, np.cos(x), color='red')
axs[1].set_title('Cos(x)')
plt.suptitle('Subplots Example')
plt.show()
```

## 9. Annotations
Add annotations to highlight specific points on a plot.

```python
# 9: Annotations
plt.figure(figsize=(6, 4))
plt.plot(x, y)
plt.annotate('Peak', xy=(np.pi/2, 1), xytext=(2, 1.2),
             arrowprops=dict(facecolor='black', shrink=0.05))
plt.title('Annotation Example')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.show()
```

## 10. Saving Figures
Save a plot as an image file using Matplotlib.

```python
# 10: Saving figures
plt.figure(figsize=(6, 4))
plt.plot(x, y)
plt.title('Saving Figure Example')
plt.savefig('line_plot_example.png') # Saves the figure to a file
plt.close()
```

## 11. Pie Chart
Create a pie chart to show proportions of categories.

```python
# 11: Pie chart
sizes = [15, 30, 45, 10]
labels = ['A', 'B', 'C', 'D']
plt.figure(figsize=(5, 5))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Pie Chart Example')
plt.show()
```

## 12. Histogram
Visualize the distribution of data using a histogram.

```python
# 12: Histogram
data_hist = np.random.randn(1000)
plt.figure(figsize=(6, 4))
plt.hist(data_hist, bins=30, color='skyblue', edgecolor='black')
plt.title('Histogram Example')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

## 13. Error Bars
Add error bars to a plot to show variability.

```python
# 13: Error bars
x_err = np.arange(0, 10, 1)
y_err = np.exp(-x_err/3)
errors = 0.1 + 0.2 * x_err
plt.figure(figsize=(6, 4))
plt.errorbar(x_err, y_err, yerr=errors, fmt='o', color='purple', ecolor='gray', capsize=5)
plt.title('Error Bars Example')
plt.xlabel('x')
plt.ylabel('exp(-x/3)')
plt.show()
```

## 14. Log Scale
Use a logarithmic scale for the y-axis in a plot.

```python
# 14: Log scale
x_log = np.linspace(0.1, 10, 100)
y_log = np.exp(x_log)
plt.figure(figsize=(6, 4))
plt.plot(x_log, y_log)
plt.yscale('log')
plt.title('Log Scale Example')
plt.xlabel('x')
plt.ylabel('exp(x) (log scale)')
plt.show()
```

## 15. 3D Plotting
Create a 3D surface plot using Matplotlib.

```python
# 15: 3D plotting
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(7, 5))
ax = fig.add_subplot(111, projection='3d')
x3d = np.linspace(-5, 5, 100)
y3d = np.linspace(-5, 5, 100)
X3d, Y3d = np.meshgrid(x3d, y3d)
Z3d = np.sin(np.sqrt(X3d**2 + Y3d**2))
ax.plot_surface(X3d, Y3d, Z3d, cmap='viridis')
ax.set_title('3D Surface Plot')
plt.show()
```

## 16. Seaborn Pairplot
Visualize pairwise relationships in a dataset using Seaborn's pairplot.

```python
# 16: Seaborn pairplot
import pandas as pd
iris = sns.load_dataset('iris')
sns.pairplot(iris, hue='species')
plt.suptitle('Seaborn Pairplot Example', y=1.02)
plt.show()
```

## 17. Seaborn Heatmap
Display a correlation matrix as a heatmap using Seaborn.

```python
# 17: Seaborn heatmap
corr = iris.corr(numeric_only=True)
plt.figure(figsize=(6, 4))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Seaborn Heatmap Example')
plt.show()
```

## 18. Boxplot and Violinplot
Compare distributions across categories using boxplots and violinplots.

```python
# 18: Seaborn boxplot and violinplot
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.boxplot(x='species', y='sepal_length', data=iris)
plt.title('Boxplot')
plt.subplot(1, 2, 2)
sns.violinplot(x='species', y='sepal_length', data=iris)
plt.title('Violinplot')
plt.tight_layout()
plt.show()
```

## 19. Swarmplot and Stripplot
Visualize individual data points with swarmplots and stripplots.

```python
# 19: Seaborn swarmplot and stripplot
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.swarmplot(x='species', y='sepal_length', data=iris)
plt.title('Swarmplot')
plt.subplot(1, 2, 2)
sns.stripplot(x='species', y='sepal_length', data=iris, jitter=True)
plt.title('Stripplot')
plt.tight_layout()
plt.show()
```

## 20. Seaborn FacetGrid
Create multiple plots based on a categorical variable using FacetGrid.

```python
# 20: Seaborn FacetGrid
g = sns.FacetGrid(iris, col='species')
g.map_dataframe(sns.histplot, x='sepal_length')
g.fig.suptitle('FacetGrid Example', y=1.05)
plt.show()
```

## 21. Seaborn Regression Plot
Visualize a linear relationship with a regression line using Seaborn.

```python
# 21: Seaborn regression plot
plt.figure(figsize=(6, 4))
sns.regplot(x='sepal_length', y='petal_length', data=iris)
plt.title('Seaborn Regression Plot')
plt.show()
```
