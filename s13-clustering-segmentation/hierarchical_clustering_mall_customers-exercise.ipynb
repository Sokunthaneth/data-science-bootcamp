## Dendrogram for Optimal Number of Clusters
The dendrogram helps visualize the hierarchical clustering process and determine the optimal number of clusters.

```python
# Plot the Dendrogram
```

From the dendrogram, you can choose the number of clusters by finding the largest vertical distance (vertical line) that you can cut without crossing any horizontal line. For this dataset, 5 clusters is a common choice.

### How to select the optimal number of clusters from the dendrogram:
1. **Look for the longest vertical line** that is not crossed by any horizontal line. This represents the largest distance between clusters before they are merged.
2. **Draw a horizontal line** across the dendrogram at a height just before the largest vertical gap. This line will intersect the dendrogram at a certain number of vertical lines.
3. **Count the number of vertical lines** the horizontal line crosses. This number is the optimal number of clusters.

In the dendrogram above, if you draw a horizontal line at a height around 200 (just before the largest jump in distance), it will cross 5 vertical lines. This means the data can be optimally grouped into 5 clusters before the next largest merge would combine very different groups together.

This method helps prevent over-segmentation (too many clusters) or under-segmentation (too few clusters), and is a visual way to determine the best cluster count for hierarchical clustering.
